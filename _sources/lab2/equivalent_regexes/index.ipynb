# Эквивалентные регулярные выражения

Рассматриваются регулярные выражения, распознающие числа в троичной системе счисления, кратные двум.

```python
import re
from time import time
import matplotlib.pyplot as plt

def match_time(regex, string):
    start = time()
    re.fullmatch(regex, string)
    return time() - start

regexes = {
    'Academic regex': '0|1(0|2)*1(1(0|2)*1|0|2)*|2(1(0|2)*1|0|2)*',
    'Negation regex': '[^12]|[^02][^1]*[^02]([^02][^1]*[^02]|[^1])*|[^01]([^02][^1]*[^02]|[^1])*',
    'Lazy regex': '0|1(0|2)*?1(1(0|2)*?1|0|2)*?|2(1(0|2)*?1|0|2)*?',

}

for regex in regexes:
    with open(f'tests.txt') as f:
        x = []; y = []
        for test in f.read().split('\n'):
            x.append(len(test))
            y.append(match_time(regexes[regex], test))
        plt.plot(x, y)

plt.legend(regexes.keys())

plt.xlabel('Length')
plt.ylabel('Milliseconds')

plt.show()
```
